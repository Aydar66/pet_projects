```python
import numpy as np
from statsmodels.formula.api import logit, probit
from statsmodels.formula.api import ols

#подключение f-распределния
from scipy.stats import f

# Не показывать FutureWarnings
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
# Не показывать ValueWarning, ConvergenceWarning из statsmodels
from statsmodels.tools.sm_exceptions import ValueWarning, ConvergenceWarning
warnings.simplefilter('ignore', category=ValueWarning)
warnings.simplefilter('ignore', category=ConvergenceWarning)
```

```python
# mod = logit(formula = , data = df)
# mod = probit(formula = , data = df)
mod = ols(formula = 'approve~unem+male+yjob+self', data = df)
res = mod.fit(cov_type='HC3')
```

```python
res.summary()
# берем p-значение = 0.153 = 15.3% > альфа (уровень значимости) = 10%, не отвергается -> значит регрессия не значима -> 
# модель не надо использовать для дальнейших исследований
```

```python
from scipy import stats
# res.fvalue, res.f_pvalue
# alpha=0.01
#это f-статистика и p-значение
t_critical = stats.t.ppf(1 - 0.01/2, 1971-7)
t_critical
```

```python
#вычислить критическое значение
#распределение Фишера
# степени свободы (две)
# для второй надо посчитать кол-во наблюдений в подгонке
#первый аргумент : 1-альфа
f.ppf(1-0.1, dfn=res.df_model, dfd=res.df_resid).round(3)
res.summary()
```

```python
# mod = logit(formula = , data = df)
# mod = probit(formula = , data = df)
mod = ols(formula = 'approve~1+appinc+np.square(appinc)+mortno+unem+dep+male+married+yjob+self', data = df)
res = mod.fit(cov_type='HC3')
```

```python
res.params
```

```python
res.wald_test('appinc, np.square(appinc)', use_f=True)
#p = 0.052
#p>alfa -> гипотеза не отвергается -> влияние незначимо
```

```python
# посчитать критическое значение
#2 из df_num
f.ppf(1-0.05, dfn=2, dfd=res.df_resid)
```

```python
import pandas as pd
import numpy as np
from statsmodels.formula.api import ols
```

```python
# Загрузка данных
df = pd.read_csv('loanapp.csv')
```

```python
# Спецификация LPM модели
mod = ols(formula='approve ~ mortno + unem + dep + male + married + yjob + self', data=df)

# Подгонка модели
# Подгонка модели осуществляется с помощью метода fit(). Здесь используем cov_type="HC3" для получения робастных стандартных ошибок.
res = mod.fit(cov_type="HC3")

# Получение коэффициентов модели
coefficients = res.params.round(3)
coefficients
```

```python
res.summary()
```

```python
f_stat = res.fvalue
f_p_value = res.f_pvalue
print(f"F-statistic: {f_stat}, P-value: {f_p_value}")
```
